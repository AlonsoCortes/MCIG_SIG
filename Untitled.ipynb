# Agrupamiento, agrupamiento espacial y geodemograf√≠a 

```python
%matplotlib inline

import seaborn as sns
import pandas as pd
import pysal as ps
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import contextily as ctx
from sklearn import cluster
```

```python
#Carga los datos
abb = gpd.read_file('DATA/viviendas.geojson')
```

```python
# Transforma a Web Mercator CRS y grafica 
ax = abb.to_crs(epsg=3857).plot(figsize=(9, 9), alpha=0.75, color='red', edgecolor='k', lw= .5);
# Agrega mapa base
ctx.add_basemap(ax)
```

```python
abb.info()
```

```python
carac = ['M2_CONST','PRECIO','PRECIO_M2', 'D_METRO', 'D_METROBUS', 'VIAL_PRIM']
```

```python
# Crea figura y ejes (en este caso es un arreglo de 2 x 3)
f, axs = plt.subplots(nrows=2, ncols=3, figsize=(12, 12))
# Hace los ejes accesibles con una sola indexaci√≥n
axs = axs.flatten()
# Comienza el loop sobre las variables de inter√©s
for i, col in enumerate(carac):
    # Selecciona el eje en el que el ir√° el mapa.
    ax = axs[i]
    # Grafica el mapa
    abb.to_crs(epsg=3857).plot(column=col, ax=ax, scheme='Quantiles', \
             linewidth=0, cmap='coolwarm', alpha=0.75)
    # Remueve "basura" del eje
    ax.set_axis_off()
    # Le pone el nombre de la variable a cada eje
    ax.set_title(col)
    ctx.add_basemap(ax)
# Despliega la figura
plt.show()
```

```python
_ = sns.pairplot(abb[carac], kind='reg', diag_kind='auto')
```

## Kmeans

```python
kmeans5 = cluster.KMeans(n_clusters=5)
kmeans5
```

```python
# Esta l√≠nea se utiliza para obtener el mismo resultado siempre que se replique el an√°lisis
np.random.seed(1234)
# Ejecuta el algoritmo de agrupamiento
k5cls = kmeans5.fit(abb[carac])
```

```python
k5cls.labels_
```

```python
abb['k5cls'] = k5cls.labels_
```

```python
# Ajusta figura y ax
f, ax = plt.subplots(1, figsize=(9, 9))
# Graficar valores √∫nicos coropletas incluyendo una leyenda y sin l√≠neas de l√≠mite
abb.to_crs(epsg=3857).plot(column='k5cls', categorical=True, legend=True, linewidth=0, ax=ax, cmap='Paired')
# Remueve ejes
ax.set_axis_off()
# Mentiene los ejes proporcionados
plt.axis('equal')
# Agrega t√≠tulo
plt.title('AirBnb Geodemographic classification for Inner London')

ctx.add_basemap(ax)
# Despliega el mapa
plt.show()
```

```python
k5sizes = abb.groupby('k5cls').size()
k5sizes
```

```python
_ = k5sizes.plot.bar()
```

```python
k5means = abb.groupby('k5cls')[carac].mean().round(2)
# Muestra la tabla transpuesta
k5means.T
```

```python
# Calcular el summary 
k5desc = abb.groupby('k5cls')[carac].describe().round(2)

# Muestra la tabla
k5desc
```

```python
# Nombrar (index) los renglones por la categor√≠a a la que pertenecen 
to_plot = abb.set_index('k5cls')
# Subconjutno que solo mantiene las variables utilizadas por el K-Means
to_plot = to_plot[carac]
# Muestra la tabla
to_plot.head()
```

```python
to_plot = to_plot.stack()
to_plot
```

```python
to_plot = to_plot.reset_index()
to_plot.head()
```

```python
to_plot = to_plot.rename(columns={'level_1': 'Rating', 0: 'Values'})
to_plot.head()
```

```python
# Setup the facets
facets = sns.FacetGrid(data=to_plot, row='Rating', hue='k5cls', \
                  sharey=False, sharex=False, aspect=2)
# Build the plot as a `sns.kdeplot`
_ = facets.map(sns.kdeplot, 'Values', shade=True).add_legend()
```

### K nearest neighbor weights

```python
w = ps.weights.KNN.from_dataframe(abb,k=5, p=2)
w
```

```python
sagg13 = cluster.AgglomerativeClustering(n_clusters=13, connectivity=w.sparse)
sagg13
```

```python
# This line is required to obtain the same results always
np.random.seed(1234)
# Run the clustering algorithm
sagg13cls = sagg13.fit(abb[carac])
```

```python
abb['sagg13cls'] = sagg13cls.labels_
```

```python
# Ajusta figura y ejes
f, ax = plt.subplots(1, figsize=(9, 9))
# Grafica por coro¬¥pletas
abb.plot(column='sagg13cls', categorical=True, legend=True, linewidth=0, ax=ax)
# Remueve ejes
ax.set_axis_off()
# mantiene los ejes proporcionados
plt.axis('equal')
#AGrega t√≠tulo
plt.title('AirBnb-based boroughs for Inner London')
# Muestra el mapo
plt.show()
```
